```python
import altair as alt
from vega_datasets import data
```

```python
cars = data.cars()
```

```python
alt.Chart(cars).mark_point().encode(
x='Horsepower',
y='Miles_per_Gallon',
color='Origin',)
```

```python
alt.Chart(cars).mark_circle().encode(
x='Horsepower',
y='Miles_per_Gallon',
color='Origin',).interactive()
```

```python
brush=alt.selection_interval()
circles=alt.Chart(cars).mark_circle().encode(
        x='Horsepower:Q',
        y='Miles_per_Gallon:Q',
        color=alt.condition(brush, 'Origin:N', alt.value('lightgray'))).add_selection(brush)
circles
```

```python
bars=alt.Chart(cars).mark_bar().encode(
    y='Origin:N',
    color='Origin:N',
    x='mean(Horsepower):Q')
bars
```

```python
brush=alt.selection_interval()
circles=alt.Chart(cars).mark_circle().encode(
        x='Horsepower:Q',
        y='Miles_per_Gallon:Q',
        color=alt.condition(brush, 'Origin:N', alt.value('lightgray'))).add_selection(brush)
bars=alt.Chart(cars).mark_bar().encode(
    y='Origin:N',
    color='Origin:N',
    x='mean(Horsepower):Q').transform_filter(brush)
circles&bars
```
